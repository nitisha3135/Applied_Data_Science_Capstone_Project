## Task 2: Mark the success/failed launches for each site on the map


Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, 
# e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
for index, record in spacex_df.iterrows():
    # TODO: Create and add a Marker cluster to the site map
    # marker = folium.Marker(...)
    marker = folium.Marker(location=[record['Lat'], record['Long']], icon=folium.Icon(color='white', icon_color=record['marker_color']))
    marker_cluster.add_child(marker)

site_map.save("my_map.html")
site_map
```

## TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians
import pandas as pd 

# Dummy launch_sites_df for demonstration if run in isolation
launch_sites_df = pd.DataFrame({
    'Launch Site': ['KSC LC-39A', 'CCAFS SLC-40', 'VAFB SLC-4E', 'CCAFS SEC 40'],
    'Lat': [28.57325, 28.563197, 34.632834, 28.563197],
    'Long': [-80.64684, -80.57682, -120.610745, -80.57682]
})

# Dummy site_map for demonstration if run in isolation
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Function to calculate distance (from previous notebook section)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# --- Approximate proximity coordinates for each launch site ---

# KSC LC-39A (Kennedy Space Center Launch Complex 39A)
ksc_lc39a_city = [28.5375, -80.6845] # Titusvill
ksc_lc39a_railway = [28.5721, -80.6545] # Near the VAB
ksc_lc39a_highway = [28.525, -80.686] # FL-405

# CCAFS SLC-40 (Cape Canaveral Space Force Station Space Launch Complex 40)
ccafs_slc40_city = [28.3922, -80.6077] # Cocoa Beach
ccafs_slc40_railway = [28.5721, -80.6545] # Near the VAB (same as KSC)
ccafs_slc40_highway = [28.3922, -80.6077] # FL-A1A

# VAFB SLC-4E (Vandenberg Space Force Base Space Launch Complex 4E)
vafb_slc4e_city = [34.6391, -120.4579] # Lompoc
vafb_slc4e_railway = [34.6338, -120.623] # Near the launch complex
vafb_slc4e_highway = [34.673, -120.590] # CA-1

# CCAFS SEC 40 (Cape Canaveral Space Force Station Space Launch Complex 40 - older name/entry, using same as SLC-40 if distinct)
ccafs_sec40_city = [28.3922, -80.6077]
ccafs_sec40_railway = [28.5721, -80.6545]
ccafs_sec40_highway = [28.3922, -80.6077]


# --- Code for creating markers and lines for city, railway, highway ---
# This loop iterates through each launch site and adds markers/lines for proximities.
for index, launch_site in launch_sites_df.iterrows():
    launch_site_name = launch_site['Launch Site']
    launch_site_lat = launch_site['Lat']
    launch_site_lon = launch_site['Long']
    launch_site_coord = [launch_site_lat, launch_site_lon]

    # Initialize proximity coordinates based on launch site name
    city_coord = None
    railway_coord = None
    highway_coord = None

    if launch_site_name == 'KSC LC-39A':
        city_coord = ksc_lc39a_city
        railway_coord = ksc_lc39a_railway
        highway_coord = ksc_lc39a_highway
    elif launch_site_name == 'CCAFS SLC-40':
        city_coord = ccafs_slc40_city
        railway_coord = ccafs_slc40_railway
        highway_coord = ccafs_slc40_highway
    elif launch_site_name == 'VAFB SLC-4E':
        city_coord = vafb_slc4e_city
        railway_coord = vafb_slc4e_railway
        highway_coord = vafb_slc4e_highway
    elif launch_site_name == 'CCAFS SEC 40':
        city_coord = ccafs_sec40_city
        railway_coord = ccafs_sec40_railway
        highway_coord = ccafs_sec40_highway
    else:
        print(f"Warning: Proximity coordinates not defined for {launch_site_name}. Skipping.")
        continue 

    # --- City ---
    if city_coord:
        distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_coord[0], city_coord[1])
        folium.Marker(
            city_coord,
            icon=DivIcon(
                icon_size=(150,20),
                icon_anchor=(0,0),
                html=f'<div style="font-size: 12px; color:blue;"><b>{launch_site_name} to City: {distance_city:10.2f} KM</b></div>',
            )
        ).add_to(site_map)
        folium.PolyLine(locations=[launch_site_coord, city_coord], weight=2, color='blue').add_to(site_map)

    # --- Railway ---
    if railway_coord:
        distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_coord[0], railway_coord[1])
        folium.Marker(
            railway_coord,
            icon=DivIcon(
                icon_size=(150,20),
                icon_anchor=(0,0),
                html=f'<div style="font-size: 12px; color:green;"><b>{launch_site_name} to Rail: {distance_railway:10.2f} KM</b></div>',
            )
        ).add_to(site_map)
        folium.PolyLine(locations=[launch_site_coord, railway_coord], weight=2, color='green').add_to(site_map)

    # --- Highway ---
    if highway_coord:
        distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_coord[0], highway_coord[1])
        folium.Marker(
            highway_coord,
            icon=DivIcon(
                icon_size=(150,20),
                icon_anchor=(0,0),
                html=f'<div style="font-size: 12px; color:orange;"><b>{launch_site_name} to Hwy: {distance_highway:10.2f} KM</b></div>',
            )
        ).add_to(site_map)
        folium.PolyLine(locations=[launch_site_coord, highway_coord], weight=2, color='orange').add_to(site_map)

site_map.save("my_map2.html")
site_map
```

#### Are launch sites in close proximity to railways?

It's difficult to definitively say "close proximity" without specific distance values in the notebook. However, visually, some launch sites appear to have railways nearby. For example, in the full solution I provided, coordinates for railways were included and it can be observed on the map.
Explanation: Launch sites often require the transport of large rocket components and fuel. Railways can be an efficient way to move these materials, so having rail access is advantageous.

#### Are launch sites in close proximity to highways?

Similar to railways, launch sites do seem to be located near highways. Again, the notebook encourages visual confirmation. The full solution shows the highways as well.
Explanation: Highways provide crucial road transportation for personnel, equipment, and supplies. Easy access to highways is important for the daily operations of a launch facility.

#### Are launch sites in close proximity to the coastline?

Yes, all launch sites are very close to the coastline. This is the most evident observation from the map. The notebook emphasizes this.
Explanation:
Safety: Launching over the ocean minimizes the risk to populated areas in case of a launch failure. Debris would fall into the water.
Trajectory: Many rockets are launched into orbits that require an eastward trajectory. Launching from the coast provides an initial velocity boost from the Earth's rotation (which is maximum near the equator) and allows for a clear path over the ocean.

#### Do launch sites keep a certain distance away from cities?

Launch sites are generally located some distance away from major cities, but not extremely far. They aren't in the middle of dense urban areas.
Explanation:
Safety: Even with launches directed over the ocean, there's a need to maintain a buffer zone from populated centers for safety.
Noise and Disturbance: Rocket launches are noisy and can cause vibrations. Being away from cities reduces disturbance.
Regulations: There are often regulations that dictate how far launch sites must be from populated areas.
